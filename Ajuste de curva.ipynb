```python
import numpy as np
from scipy.optimize import curve_fit
import matplotlib.pyplot as plt

# Define la función que deseas ajustar
def funcion(x, a, b, c):
    return a * np.exp(-b * x) + c

# Genera algunos datos de ejemplo
x = np.array([2014, 2015, 2016, 2017, 2018, 2019, 2020])
y = np.array([16.95, 16.45, 16.20, 16.21, 16.19, 15.81, 15.39])

# Agrega ruido a los datos
np.random.seed(0)
y_ruidoso = y + 0.2 * np.random.normal(size=len(x))

# Realiza el ajuste de curva
parametros_optimizados, matriz_covarianza = curve_fit(funcion, x, y_ruidoso)

# Grafica los datos y la curva ajustada
plt.scatter(x, y_ruidoso, label='Datos ruidosos')
plt.plot(x, funcion(x, *parametros_optimizados), 'r-', label='Curva ajustada')
plt.legend()

# Muestra la gráfica
plt.show()

```
